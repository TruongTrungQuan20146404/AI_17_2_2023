```python
#BT_Machine learning

import pandas as pd 
url = 'classification.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation,annot = True)
```

```python
data.plot(kind='scatter',x='interest',y='success',figsize=(9,9), color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['interest']),y=data['success'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
```

```python
model.score(X = pd.DataFrame(data['interest']),y=data['success'])
```

```python
data.plot(kind='scatter',x='interest',y='success',figsize=(9,9), color='black')
y_pred=model.predict(X = pd.DataFrame(data['interest']))
plt.plot(data['interest'],y_pred, color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['interest'],data['interest']**2]).T
model.fit(X,y=data['success'])
print('He so Intercept la:', model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y=data['success']))
```

```python
import numpy as np
range = np.arange(10,80,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter',x='interest',y='success',figsize=(9,9),color='black', xlim=(0.7))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['interest'],data['interest']**2,data['interest']**3]).T
model.fit(X,y=data['success'])
print('He so Intercept la:', model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y=data['success']))
```

```python
import numpy as np


range = np.arange(10,80,0.1)
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter',x='interest',y='success',figsize=(9,9),color='black', xlim=(0.7))
plt.plot(range,y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['interest'],data['interest']**2,data['interest']**3,data['interest']**4,data['interest']**5,data['interest']**6,data['interest']**7]).T
model.fit(X,y=data['success'])
print('He so Intercept la:', model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:', model.score(X,y=data['success']))
```
